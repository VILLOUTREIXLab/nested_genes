#### random pairs

```python
def rand_pair_diff(raw,nb=567):
    pair_diff=pd.DataFrame()
    g1=[]
    g2=[]
    while len(g1)<nb:
        r1=random.randint(0,len(raw)-1)
        r2=random.randint(0,len(raw)-1)
        if raw.at[r1,"Chromosome/scaffold name"]!=raw.at[r2,"Chromosome/scaffold name"]:
            g1.append(raw.at[r1,"Gene stable ID"])
            g2.append(raw.at[r2,"Gene stable ID"])
        
    pair_diff["GeneID1"]=g1
    pair_diff["GeneID2"]=g2
    return pair_diff    
# df=rand_pair_diff(raw,nb=10)
```

#### nested pairs

```python
### select nested pairs

def nested_pair(raw, name=False):
    pair_nested=pd.DataFrame()
    g1=[]
    g2=[]

    # separation by chr
    chI=raw[raw["Chromosome/scaffold name"]=="I"]
    chII=raw[raw["Chromosome/scaffold name"]=="II"]
    chIII=raw[raw["Chromosome/scaffold name"]=="III"]
    chIV=raw[raw["Chromosome/scaffold name"]=="IV"]
    chV=raw[raw["Chromosome/scaffold name"]=="V"]
    chX=raw[raw["Chromosome/scaffold name"]=="X"]
    chMtDNA=raw[raw["Chromosome/scaffold name"]=="MtDNA"]
    #list for iteration
    ls_chr=[chI,chII,chIII,chIV,chV,chX,chMtDNA]
    
    for ch in ls_chr:    
        # since we only interessed in 1/-1 nested paire
        strandpos=ch[ch["Strand"]==1]
        strandneg=ch[ch["Strand"]==-1]  
        
        for i,k in strandpos.iloc[:,4:6].iterrows():
            #start and end of the potential host gene
            st,en=k["Gene start (bp)"],k["Gene end (bp)"]
            
            #listing every gene that start after and end before the host
            l1=strandneg.iloc[:,4] > st
            l2=strandneg.iloc[:,5] < en
            
            group=pd.concat([l1,l2],axis=1)
            group['both']=group.sum(axis=1)
            
            ls_id=group[group['both']==2].index.tolist()
            
            #for different verbose
            if name:
                # for mutilpe nested in the same host
                for j in ls_id:
                    g1.append(ch.loc[i,'Gene name'])
                    g2.append(ch.loc[j,'Gene name'])
            else:
                for j in ls_id:
                    g1.append(ch.loc[i,'Gene stable ID'])
                    g2.append(ch.loc[j,'Gene stable ID'])
                
        # same but looking for host in -1 strand            
        for i,k in strandneg.iloc[:,4:6].iterrows():
            #start and end of the potential host gene
            st,en=k["Gene start (bp)"],k["Gene end (bp)"]
        
            #listing every gene that start after and end before the host
            l1=strandpos.iloc[:,4] > st
            l2=strandpos.iloc[:,5] < en
            
            group=pd.concat([l1,l2],axis=1)
            group['both']=group.sum(axis=1)
            
            ls_id=group[group['both']==2].index.tolist()
            
            if name:
                # for mutilpe nested in the same host
                for j in ls_id:
                    g1.append(ch.loc[i,'Gene name'])
                    g2.append(ch.loc[j,'Gene name'])
            else:
                for j in ls_id:
                    g1.append(ch.loc[i,'Gene stable ID'])
                    g2.append(ch.loc[j,'Gene stable ID'])
        
    pair_nested["Host"]=g1
    pair_nested["Nested"]=g2
    return pair_nested.drop_duplicates(ignore_index=True)
    
#df=nested_pair(raw,name=(True))
print(df)
#df.to_csv('nested_pair_name.csv')
```

#### get expression value linked to pairs

```python
def nb_cell_expression(df, pl, name=False):

    #init the output dataframe
    nb_cell_exp=pd.DataFrame(columns=['Host_gene','Nested_gene','exp_both','exp_U','ratio_coexp','exp_host','exp_nested'])

    #Choose whether to use gene name or Whormbase ID 
    if name:
        marker=df['gene'].drop_duplicates()
        n_cell=len(df['lineage'].drop_duplicates()) #how many cell are in the data
    else:
        marker=df['gene_id'].drop_duplicates()
        n_cell=len(df['lineage'].drop_duplicates())

    # for each gene in each pair get the number of cell where tpm is above 0 
    for k in pl.itertuples():
        if len(marker[marker==k[1]]) > 0:            
            i1=marker[marker==k[1]].index[0]
            g1=df.iloc[ i1: i1+n_cell, : ].reset_index(drop=True)['tpm'] > 0
        else :
            g1=pd.Series([],dtype=bool) #to avoid conflict when the gene is not presente in expression data
        
        if len(marker[marker==k[2]]) > 0:            
            i2=marker[marker==k[2]].index[0]
            g2=df.iloc[ i2 : i2+n_cell, : ].reset_index(drop=True)['tpm'] > 0
        else :
            g2=pd.Series([],dtype=bool)

        # look in how many cell both gene are expressed
        comb=pd.concat([g1,g2],axis=1,keys=[k[1],k[2]])
        comb['both']=comb.sum(axis=1)
        
        nb_cell_exp.at[k[0],'Host_gene'] = k[1]
        nb_cell_exp.at[k[0],'Nested_gene'] = k[2]
        nb_cell_exp.at[k[0],'exp_both'] = (comb['both']==2).sum() # get gene1 inter gene2
        nb_cell_exp.at[k[0],'exp_U'] = g1.sum() + g2.sum() - nb_cell_exp.at[k[0],'exp_both'] # get gene1 union gene2
        if nb_cell_exp.at[k[0],'exp_U'] == 0:
            nb_cell_exp.at[k[0],'ratio_coexp'] = 0 # to avoid divid by 0
        else:
            # ratio coexp = number of cell that express both divide by number of cell tha express at least one gene
            nb_cell_exp.at[k[0],'ratio_coexp'] = nb_cell_exp.at[k[0],'exp_both'] / (nb_cell_exp.at[k[0],'exp_U']) 
        nb_cell_exp.at[k[0],'exp_host'] = g1.sum()
        nb_cell_exp.at[k[0],'exp_nested'] = g2.sum()

    return nb_cell_exp

# nb_cell_expression(exp7)
# nb_cell_expression(exp8)
```

#### iteration random and close pairs

```python
### two way for caluculation of outlier that will be used later

def outlier_up_zscore(ls):
    z=stats.zscore(ls)
    return sum(ls>3)

def outlier_up_IQR(ls):
    Q1=np.quantile(ls,.25)
    Q3=np.quantile(ls,.75)
    IQR=Q3-Q1
    lim_up=Q3+(1.5*IQR)
    return sum(ls>lim_up)
```

```python
def list_exp_outlier_random(raw, exp, pairs='diff', it=1000, nb=567):
    ls_mean_exp=[]
    ls_exp_both=[]
    ls_exp_U=[]
    ls_out_up_z=[]
    ls_out_up_iqr=[]
    for i in range(it):
        if pairs=='diff':
            print('it',i,'diff')
            pl = rand_pair_diff(raw, nb)
        elif pairs=='close':
            print('it',i,'close')
            pl = rand_pair_close(raw, nb)
        nb_cell_exp = nb_cell_expression(exp, pl)
        coexp=nb_cell_exp['ratio_coexp'].astype('float64')
        ls_mean_exp.append(np.mean(coexp))
        ls_exp_both.append(sum(nb_cell_exp['exp_both']>0))
        ls_exp_U.append(sum(nb_cell_exp['exp_U']>0))
        ls_out_up_z.append(outlier_up_zscore(coexp))
        ls_out_up_iqr.append(outlier_up_IQR(coexp))
    
    return pd.concat([pd.Series(ls_mean_exp,name='mean_exp'),
                      pd.Series(ls_exp_both,name='exp_both'),
                      pd.Series(ls_exp_U,name='exp_U'),
                      pd.Series(ls_out_up_z,name='out_z'),
                      pd.Series(ls_out_up_iqr,name='out_iqr')], axis=1)
```

```python
### function to get the value of outlier z1 and z3

def get_z_val(values):
    m = np.mean(values)
    sd = np.std(values)
    zd3 = m-(3*sd)
    zd1 = m-(1*sd)
    zu1 = m+(1*sd)
    zu3 = m+(3*sd)
    return zd3,zd1,zu1,zu3
```

```python
def outlier_up_IQR(ls):
    Q1=np.quantile(ls,.25)
    Q3=np.quantile(ls,.75)
    IQR=Q3-Q1
    out_up=Q3+(1.5*IQR)
    return out_up
```

```python
out_up=outlier_up_IQR(a)
print(sum(a>out_up))
out_up=outlier_up_IQR(b)
print(sum(b>out_up))
out_up=outlier_up_IQR(c)
print(sum(c>out_up))
```

Statisticians use modified Z-score to minimize the influence of outliers on Z-score. ## 4/ Probability of expressing the host gene, when the nested is expressed, is higher than the opposite

### nes/profil coexp different Ã  eclaircir

```python
profil_s8=pd.read_pickle('/home/mateo/IGE/code_nested_gene_M/celltype_profil_s8.pkl')
profil_s7=pd.read_pickle('/home/mateo/IGE/code_nested_gene_M/celltype_profil_s7.pkl')

def prob_of_expression(profil):
    profil=profil.drop(len(profil)-1)
    B=profil['propB']
    H=profil['propH']
    N=profil['propN']
    l=[]
    l2=[]
    for i in range(len(profil)):
        if B[i]+N[i] > 0:
            l.append(B[i]/(B[i]+N[i]))
        if B[i]+H[i] > 0:
            l2.append(B[i]/(B[i]+H[i]))

    fig=px.histogram(l, histnorm='percent', range_x=[0,1]).update_layout(yaxis_title="percentage of pair",xaxis_title="probability of expressing host if nested is expressed")
    fig.add_vline(x=sum(l)/len(l), line_width=2, line_color='red', annotation_text='means sum(prob) / nb prob', annotation_position='top right')
    fig.show()
    print(sum(l)/len(l), 'mean of probability host if nested expressed')

    fig2=px.histogram(l2, histnorm='percent', range_x=[0,1]).update_layout(yaxis_title="percentage of pair",xaxis_title="probability of expressing nested if host is expressed")
    fig2.add_vline(x=sum(l2)/len(l2), line_width=2, line_color='red', annotation_text='means sum(prob) / nb prob', annotation_position='top right')
    fig2.show()
    print(sum(l2)/len(l2), 'mean of probability nested if host expressed')

prob_of_expression(profil_s7)
```
